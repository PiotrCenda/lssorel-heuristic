```python
import random
import pygad
import numpy as np
from fitness import fitness_func_energy, on_generation, on_generation_xlastovka
from xlastovka import xlastovka_search
from datetime import datetime
```

```python
SEED = 1234
random.seed(SEED)
```

```python
time_start = datetime.now()
ga_entity = pygad.GA(
    num_generations=30000,
    num_parents_mating=2,
    sol_per_pop=5,
    num_genes=64,
    init_range_low=-1,
    init_range_high=2,
    gene_type=int,
    gene_space=[-1,1],
    fitness_func=fitness_func_energy,
    on_generation=on_generation_xlastovka,
    stop_criteria="reach_0.0",
    random_seed=SEED,
    crossover_type="scattered",
    mutation_type='random',
    parent_selection_type="rank",
    mutation_percent_genes=10
)

ga_entity.run()
print("Time taken: ", datetime.now() - time_start)
```
